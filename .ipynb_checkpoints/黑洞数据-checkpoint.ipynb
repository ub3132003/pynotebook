#### 于是我想了个办法

用棒棒糖来表示66亿这个数字;把他看出4个数字相乘，第一个数字看成棒棒糖的杆子，第二个看出棒棒糖的糖球的面积，第三个看出棒棒糖的颜色。这样每个数字只有285，应该能画下来了

```python
earth_t=pow(earth,1/4)
earth_color=(0,1,1)
base_sun=100000
fig, ax = plt.subplots(1,1,figsize=(10.5,7.5), facecolor='white', dpi= 80)
ax.vlines(x=1, ymin=0, ymax=earth_t, color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=1, y=earth_t, s=earth_t**2*base_sun, color=(0,1,1), alpha=0.7, cmap='inferno')

ax.vlines(x=2, ymin=0, ymax=sun, color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=2, y=sun, s=base_sun, color=(0,earth_t,earth_t), alpha=0.7)

ax.set_xticks([0,1,2,3])
ax.set_xticklabels([None,"地球","太阳"])
```

```python
hole_t=pow(hole,1/4)
sun_color=(0,1,1)
hole__color=(0,1/hole_t,1/hole_t)
base_sun=100
fig, ax = plt.subplots(1,1,figsize=(16,20), facecolor='white', dpi= 80)
ax.vlines(x=1, ymin=0, ymax=sun, color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=1, y=sun, s=sun**2*base_sun, color=sun_color, alpha=0.7,cmap='inferno')

ax.vlines(x=2, ymin=0, ymax=hole_t, color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=2, y=hole_t, s=hole_t**2*base_sun, color=hole__color, alpha=0.7, cmap='inferno')

ax.set_xticks([0,1,2,3])
ax.set_xticklabels([None,"太阳","黑洞"])
```

# 颜色映射
c=y 表示 颜色随y变化

```python
gradient = np.linspace(0, 1, 256)  
gradient = np.vstack((gradient, gradient)) 
plt.imshow(gradient, aspect='auto', cmap='inferno')
plt.cm.inferno
```

```python
gradient = np.linspace(1, 0, 256)  
xy = np.linspace(0, 1, 256)  
for i,xy in zip(gradient,xy):
    plt.scatter(xy,xy, c=(0,i,i))
plt.cm.inferno
```

```python
gradient = np.linspace(1, 0, 256)  
xy = np.linspace(0, 1, 256)  

plt.scatter(xy,xy, c=xy[::-1], cmap='inferno')
```
