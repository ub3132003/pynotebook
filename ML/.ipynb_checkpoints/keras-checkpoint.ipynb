```python
import keras
```

```python

import pandas as pd
import os
from  pathlib import Path
train_file=r"zhengqi_train.txt"
test_file=r"zhengqi_test.txt"
data_dir=r"D:\project\tianchi\steam_prodict"
"""Path(data_dir,train_file)"""
table= pd.read_table(Path(data_dir,train_file)).iloc[:,:38]

table.head()
```

```python
%matplotlib inline
v0_tar=table[["V0","target"]]
v0_tar.head()
v0_tar.plot()
```

# æ¨¡åž‹

```python
model = keras.Sequential() #çº¿æ€§æ ¸
```

```python
from keras import layers
model.add(layers.Dense(1,input_dim=38))
```

# è®­ç»ƒ

```python
model.summary()
model.compile(optimizer='adam',
             loss='mse') #ä¼˜åŒ–å‡½æ•°, mse æ–¹å·®
x=v0_tar.index
y=v0_tar[["V0"]]
model.fit(x,y,epochs=100)
```

# é¢„æµ‹

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)

plt.plot(x,model.predict(x),color="red")
```

```python
import numpy as np
import random
import pandas as pd
#åŽŸå‡½æ•°
x=np.arange(1,10,1,dtype=int)
y=1.9*x+3.3
y=list ( map( lambda n:n+random.randint(-1,1),y))
#åŠ å™ªå£°
y2=4*x+1.3
y2=list ( map( lambda n:n+random.randint(-1,1),y))
pd.DataFrame({0:y,1:y2})
```

```python
from keras import layers
import keras
model = keras.Sequential() #çº¿æ€§æ ¸
model.add(layers.Dense(1,input_dim=1))
model.compile(optimizer='adam',
             loss='mse') #ä¼˜åŒ–å‡½æ•°, mse æ–¹å·®
model.summary()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
model.fit(x,y,epochs=2000)
```

```python
plt.scatter(x,y)
plt.plot(x,1.9*x+3.3)
plt.plot(x,model.predict(x),color="red")
```
