```python
import io
from PIL import Image
import numpy as np
def generateEmbedding_iostream(imgStream,id="",imageMode=None):
    face_image1= np.array( Image.open(io.BytesIO(imgStream)))
    #face_image1=cv2.cvtColor(face_image1, cv2.COLOR_RGB2GRAY)
    if imageMode:
        face_image1=cv2.cvtColor(face_image1, imageMode)
        
    face1=faceRecognition.add_identity(face_image1,id)
    assert len(face1)==1,"more then one face detect"    
    return {"id":face1[0].name,"embedding":face1[0].embedding}
def generateEmbedding_filename(imgPath,id="",imageMode=None):
    face_image1=cv2.imread(imgPath) #**
    if imageMode:
        face_image1=cv2.cvtColor(face_image1, imageMode)
    
    face1=faceRecognition.add_identity(face_image1,id)
    assert len(face1)==1,"more then one face detect"    
    return {"id":face1[0].name,"embedding":face1[0].embedding}

with open(r"X:/share/image/mydataset/test_img/andrew/andrew10.jpg","rb") as im:
    img=im.read()
    face1=generateEmbedding_iostream(img,"L", cv2.COLOR_RGB2BGR)   

face2=generateEmbedding_filename(r"X:/share/image/mydataset/test_img/andrew/andrew10.jpg","",cv2.COLOR_RGB2BGR)
    
    
    
face1

```

## 传输编码

```python
import json

class MyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(MyEncoder, self).default(obj)
face_BGR=generateEmbedding_filename(img_path,"")
# 
js=json.dumps(face_BGR["embedding"],cls=MyEncoder)
js=json.loads(js)
np.array(js)

import pickle 
import base64
pk=pickle.dumps(face_BGR["embedding"])
b64=base64.b64encode(pk)
#b64=base64.b64decode(b64)
pkd=pk.decode("CP437")

js=json.dumps(pkd)
# #js=json.dumps(str(b64))
print(js)
js=json.loads(js)
print(js)
pickle.loads(js.encode("cp437"))
# np 2json
a=np.array([2,3,56],dtype=np.int32)
json.dumps(a,cls=MyEncoder)

#pickle.loads(base64.b64decode(js))
```
