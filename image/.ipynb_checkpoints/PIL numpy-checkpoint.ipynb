# 字节流 图片处理

```python
import io
from PIL import Image 
import matplotlib.pyplot as plt
import cv2
import numpy as np
with open('1.png', mode='rb') as im:
    img=im.read()
    
#bstream #point**********
image=Image.open(io.BytesIO(img)) # 重点1 读取io流 重点2 转np
#cvim=cv2.imread(image)
data2=np.array(image)
#plt.imshow(image)
plt.imshow(image)

data1=cv2.imread("1.png")

print(type(data1),"=============",type(data2))
```
