## cv2 转iobytes

```python
# im decode encode
# encode MAT
print(cv2.imencode("1.png",mat))
imgcode=cv2.imencode("1.png",mat)[1]
str_encode=imgcode.tostring()
# with open("1.png","rb") as f:
#     print(bytes(f.read()))

str_encode
```

```python
# decode
import numpy as np

nparr = np.frombuffer(str_encode, np.uint8)
mat = cv2.imdecode(nparr, 1)

# 打开文件流
with open ("1.png","rb") as f:
    nparr=np.frombuffer(f.read(), np.uint8)
    mat= cv2.imdecode(nparr,1)
```

```python
cv2.imshow("1",mat)# 这里调用gui  注意这几步的意义
cv2.waitKey()  #
cv2.destroyAllWindows() #
```
