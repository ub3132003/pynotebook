# 相似人脸

```python
import random
import matplotlib
import math
import pandas as pd
```

```python
# object
ALL=1000#总人数
a={"real":-1,"may":-1,"score":-1} #一个人的单次比对 结果条 [ n,1,0.9] 真人，人，可能性 # {  ，，}
A=[] #所有信息条 的集合 (a ) 
mayman=-9 #
realman=0 # 一个数字代表一个真人
manSet=set() #所有真人集合
score=0.789 # 代表一人与一个人的相似度
scoreMap= {} #代表22之间的分数组合 map集合
newOne=() #  (a) # 待测试的人集 条

#comp(1,2)  # 2个人之间比较的到的真实分数
#ImgComp(1,2) #2人的照片比较的 浮动分数
```

```python
def imgComp(a,b):
    #if a<0 or b <0 :return None 
    return scoreMap[(a,b)]+random.uniform(-0.05,0)


def comp(a,b):
    isSelf= True if a==b else False
    return randomScore(isSelf)

def randomScore( isSelf):
    if isSelf:
        return random.uniform(0.92,0.99) #同一个人 会出现的相似度s
    else:
        return random.uniform(0.29,0.92)#不同人之间会出现的相似度
```

```python
scoreMap.clear()
manSet={ i for i in range(ALL)  }

for i in manSet:
    for j in manSet:
        scoreMap[(i,j)]=   comp(i,j) if (j,i) not in scoreMap.keys() else scoreMap[(j,i)]
        
```

```python
def D( realman):
    global mayman
    a={"real":-1,"may":-1,"score":-1};maxScore=0
    for i in A:#是否有相似
       
        score=imgComp( realman, i["real"])
        
        if score> 0.92 and score>maxScore:#阈值
            maxScore=score
            a={"real":realman,"may":i["may"],"score":score}

        else:
            continue
    
    if a["real"]==-1: #没有相似 
        mayman-=1
        a={"real":realman,"may":mayman,"score":1} # 没有相似怎么办
    A.append(a)
```

```python
newTop=800-1;newLen=10000
newOne=[random.randint(0,newTop) for i in range(newLen)]
newOne=[{"real":i,"may":-1,"score":-1} for i in newOne]
```

```python
A.clear();mayman=-9
for i in newOne:
    D(i["real"])


print ("max",min( [ i["may"] for i in A] ))
```
